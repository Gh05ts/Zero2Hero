```python
N = torch.zeros((27, 27), dtype=torch.int32)
```

```python
chars = sorted(list(set(''.join(words))))
stoi = {s: i+1 for i, s in enumerate(chars)}
# stoi['<S>'] = 26
# stoi['<E>'] = 27
stoi['.'] = 0
itos = {i: s for s, i in stoi.items()}
```

```python
P = (N+1).float() # smoothening by adding 1 to avoid inf for 0 prob
P /= P.sum(1, keepdim=True)
# IMP: using broadcasting which aligns R->L
# P.shape
# P.sum(1, keepdim=True)
```

```python
g = torch.Generator(device='cuda').manual_seed(2147483647)

for i in range(5):
    out = []
    ix = 0
    while True:
        # p = N[ix].float()
        # p = p / p.sum()
        p = P[ix]
        ix = torch.multinomial(p.to('cuda'), num_samples=1, replacement=True, generator=g).item()
        out.append(itos[ix])
        if ix == 0:
            break
    print(''.join(out))
```

```python
# log(a*b*c) = log(a) + log(b) + log(c)
# hint for why log is a good function here
# GOAL: maximize likelihood of the data w.r.t. model parameters (statistical modeling)
# equivalent to maximizing the log likelihood (because log is monotonic)
# equivalent to minimizing the negative log likelihood
# equivalent to minimizing the average negative log Likelihood
```

```python
log_likelihood = 0.0
n = 0 # for averagin nll
for w in words[:3]:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        prob = P[ix1, ix2]
        logprob = torch.log(prob)
        log_likelihood += logprob
        n += 1
        print(f'{ch1}{ch2}: {prob: .4f} {logprob: .4f}')

print(f'{log_likelihood=}')
nll = -log_likelihood
print(f'{nll=}')
print(f'{nll/n}')
```

```python
# create the dataset
xs, ys = [], []
for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        xs.append(ix1)
        ys.append(ix2)

xs = torch.tensor(xs, device='cuda')
ys = torch.tensor(ys, device='cuda')
num = xs.nelement()
print('num of examples: ', num)

g = torch.Generator(device='cuda').manual_seed(2147483647)
W = torch.randn((27, 27), generator=g, device='cuda', requires_grad=True)
```

```python
(W ** 2).mean()
```

```python
# gradient descent
for k in range(10):
    # forward pass
    xenc = F.one_hot(xs, num_classes=27).float() # input to network: one-hot encoding
    logits = xenc @ W # predict log-counts
    counts = logits.exp() # counts, equiv to N
    probs = counts / counts.sum(1, keepdims=True)
    loss = -probs[torch.arange(num, device='cuda'), ys].log().mean() + 0.01 * (W**2).mean() # regularization/ smoothening, pushing W towards 0
    print(loss.item())

    # backward pass
    W.grad = None
    loss.backward()

    # update
    W.data += -1 * W.grad # can be bigger even 50
```

```python
# This final value 2.4604601860046387 is very close to the initial 2.4248201847076416 which we got by manual counting in the beginning
# However the initial approach was not scalable for further updates where we would like to update the logic to take bigger words into
# consideration rather than just bigrams as it would become too large too quickly, while this approach will scale very easily

# sample From the "neurat net model
g = torch.Generator(device='cuda').manual_seed(2147483647)
for i in range(5):
    out = []
    ix = 0
    while True:
        # BEFORE:
        # p = P[ix]

        # NOW:
        xenc = F.one_hot(torch.tensor([ix], device='cuda'), num_classes=27).float()
        logits = xenc @ W # predict log-counts
        counts = logits.exp() # counts, equivalent to N
        p = counts / counts.sum(1, keepdims=True) # probabilities for next character
        
        ix = torch.multinomial(p.to('cuda'), num_samples=1, replacement=True, generator=g).item()
        out.append(itos[ix])
        if ix == 0:
            break
    print(''.join(out))
```
